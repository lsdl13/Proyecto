Aplicamos la función valuecounts para detectar cuantos valores repetidos de países hay en todas las filas del dataset. Tras ello, seleccionamos mediante la función index aquellos valores que se hayen repetidos 5 veces o menos y eliminamos las filas que no contengan un país que este incluido 5 veces entre todos los países. Escogemos el valor 5 equivalente a un caso cada 50 años. 

```python
value_counts = df_Country['Country'].value_counts()
remote_cases = value_counts[value_counts <= 5].index
df_Country = df_Country[~df_Country.Country.isin(remote_cases)]
df_Country["Country"].value_counts()
```
