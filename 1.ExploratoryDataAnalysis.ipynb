# 1.2. Limpieza preliminar de la columna Fatal?

Detectamos que la columna Fatal? tiene algunos valores que merece la pena uniformizar para un posterior análisis y procedemos a ello.

```python
df['Fatal?'] = df['Fatal?'].str.strip()
df['Fatal?'] = df['Fatal?'].fillna('U')
df['Fatal?'] = df['Fatal?'].str.replace('n', 'N')
df['Fatal?'] = df['Fatal?'].str.replace('#VALUE!', 'U')
df['Fatal?'] = df['Fatal?'].str.replace('F', 'Y')
df['Fatal?'] = df['Fatal?'].str.replace('UNKNOWN', 'U')
```
